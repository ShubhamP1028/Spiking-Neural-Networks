In computational neuroscience, biological neuron models are typically classified by their biological realism and mathematical complexity. Models range from highly detailed simulations of chemical and electrical processes to simplified "point-neuron" abstractions.
1. Conductance-Based (Biophysical) Models
These are the most realistic models, simulating the physical flow of ions (sodium, potassium, etc.) across the cell membrane.
-  Hodgkin-Huxley (H-H) Model: a foundational computational model in neuroscience that describes how action potentials (nerve impulses) are initiated and propagated, based on squid giant axon experiments.

image.png

 it uses differential equations to describe how individual ion channels open and close to generate action potentials - $$\frac{dV}{dt} = \frac{1}{C_m} \left( I_{inj}(t) - \bar{g}_K n^4 (V - E_K) - \bar{g}_{Na} m^3 h (V - E_{Na}) - \bar{g}_L (V - E_L) \right)$$

* Action Potential Generation: The model demonstrates how voltage-dependent Sodium (Na+) and Potassium (K+) conductances create the rapid positive and negative voltage spikes of an action potential.
* Four Differential Equations: The model is defined by a set of four nonlinear differential equations that track the membrane potential (V) and three gating variables (n, m, h).
* Gating Variables:

  * n (Potassium activation): Models the delayed rectifier channels.
  * m (Sodium activation): Models the fast opening of channels.
  * h (Sodium inactivation): Models the closing of channels.
* Ion Currents: The total membrane current is the sum of Na+ current (I_Na), K+ current (I_k), and a leak current (I_L)


-  Morris-Lecar Model: A simplified two-variable model originally created to describe the oscillations in barnacle muscle fibers.


-  FitzHugh-Nagumo Model: A further abstraction of the H-H model that captures the essential "excitable" nature of a neuron using only two variables.
-  Compartmental Models: These represent the physical shape of a neuron (dendrites, soma, axon) by breaking it into many small connected "compartments," each with its own H-H dynamics.

2. Artificial Neuron Models
Used primarily in AI and machine learning, these are the most abstract.
*  McCulloch-Pitts Neuron: The first artificial neuron; a simple binary threshold unit.
*  Perceptron: A linear classifier that uses a step function to produce an output.
*  Sigmoid/ReLU Units: Modern artificial neurons used in deep learning that apply continuous non-linear activation functions.
