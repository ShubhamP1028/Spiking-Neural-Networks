In computational neuroscience, biological neuron models are typically classified by their biological realism and mathematical complexity. Models range from highly detailed simulations of chemical and electrical processes to simplified "point-neuron" abstractions.
1. Conductance-Based (Biophysical) Models
These are the most realistic models, simulating the physical flow of ions (sodium, potassium, etc.) across the cell membrane.

<u><>Hodgkin-Huxley (H-H) Model</b></u>: a foundational computational model in neuroscience that describes how action potentials (nerve impulses) are initiated and propagated, based on squid giant axon experiments.

image.png

 it uses differential equations to describe how individual ion channels open and close to generate action potentials - $$\frac{dV}{dt} = \frac{1}{C_m} \left( I_{inj}(t) - \bar{g}_K n^4 (V - E_K) - \bar{g}_{Na} m^3 h (V - E_{Na}) - \bar{g}_L (V - E_L) \right)$$

---
* Action Potential Generation: The model demonstrates how voltage-dependent Sodium (Na+) and Potassium (K+) conductances create the rapid positive and negative voltage spikes of an action potential.

#### Action Potential : Na⁺ and K⁺
The spike is driven by the rapid exchange of Sodium and Potassium ions through voltage-gated channels.
* Resting State: The cell is "polarized" at about -70mV.
  * Na⁺ is concentrated outside.
  * K⁺ is concentrated inside.
* Depolarization (The Upstroke): When a stimulus hits, Na⁺ channels fly open.
Positive Sodium rushes into the cell.
The internal voltage flips from negative to positive (+30mV).
* Repolarization (The Downstroke): Na⁺ channels close, and K⁺ channels open.
Potassium rushes out of the cell.
The voltage drops back down toward its resting state.



---
<b>Four Differential Equations:</b> The model is defined by a set of four nonlinear differential equations that track the membrane potential (V) and three gating variables (n, m, h).

<b> Gating Variables:</b>

  * n (Potassium activation): Models the delayed rectifier channels.
  * m (Sodium activation): Models the fast opening of channels.
  * h (Sodium inactivation): Models the closing of channels.
* Ion Currents: The total membrane current is the sum of Na+ current (I_Na), K+ current (I_k), and a leak current (I_L)

<u><b>Morris-Lecar Model:</u></b> A simplified two-variable model originally created to describe the oscillations in barnacle muscle fibers.

<u><b>FitzHugh-Nagumo Model:</u></b> A further abstraction of the H-H model that captures the essential "excitable" nature of a neuron using only two variables.

<u><b>Compartmental Models:</u></b> These represent the physical shape of a neuron (dendrites, soma, axon) by breaking it into many small connected "compartments," each with its own H-H dynamics.

<b><u>2. Artificial Neuron Models</u></b>
Used primarily in AI and machine learning, these are the most abstract.

Generations Breakdown
- 1st Gen (1940s): McCulloch-Pitts—binary threshold logic gates, no learning.

- 2nd Gen (1980s+): ANNs—continuous activations (sigmoid/tanh/ReLU), backprop, deep hierarchies. Your perceptrons → Transformers.

- 3rd Gen (1997+): SNNs—temporal/spiking coding (info in spike timing/rates), bio-plausible rules (STDP), event-driven. IF/LIF as neurons, but network dynamics + learning make "3rd gen."
